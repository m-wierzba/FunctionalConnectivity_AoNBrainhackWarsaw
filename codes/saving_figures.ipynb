```python
import numpy as np
import matplotlib.pyplot as plt
import helpers as hel # for now provides functions figure(), labels() and savef()
import matplotlib.gridspec as gridspec # this one is only needed for example 3
```

```python
fname = 'example1'
folder_output = './'
```

```python
# create / load the data:
data = np.random.rand(10,10)
```

```python
# plot the figure:
plt.figure()
plt.imshow(data)
```

```python
# save the figure in 3 formats and 2 sizes:
plt.savefig(folder_output + fname + '_300dpi.jpg', dpi=300)
plt.savefig(folder_output + fname + '_300dpi.png', dpi=300)
plt.savefig(folder_output + fname + '_300dpi.pdf', dpi=300)
plt.savefig(folder_output + fname + '_600dpi.jpg', dpi=600)
plt.savefig(folder_output + fname + '_600dpi.png', dpi=600)
plt.savefig(folder_output + fname + '_600dpi.pdf', dpi=600)
```

# Now do similar stuff with the use of our beautiful plotting routines

```python
# to get an interactive plot:
%matplotlib nbagg
dpis = [300, 600]
formats = ['jpg', 'png', 'pdf']
```

```python
fname = 'example2'
fig, gs = hel.figure(size = [6., 6.], bottom = 0.1) # with default parameters the function returns a figure handle and grid object
ax = plt.subplot(gs[0,0])

ax.imshow(data)
hel.labels(ax, title = 'Some random data', xlabel = 'numbers from 0 to 9', ylabel = 'numbers from 9 to 0', xcolor = 'c')

hel.savef(fig, fname, dpis, formats) # now the figure should be there
```

```python
# And sth more elaborate
fname = 'example3'
fig, gs = hel.figure([1, 3], [9, 6], wspace = 0.3) # produces a grid with 1 row and 3 columns

ax1 = plt.subplot(gs[0,0])
ax1.imshow(data)
hel.labels(ax1, title = 'Again the random data', xlabel = 'numbers', ylabel = 'same numbers')

# Here is how to nest grid in a grid
subgs = gridspec.GridSpecFromSubplotSpec(2, 1, gs[0,1:], hspace = 0.35) 
# from gs take 0th row and all columns beginning with 1st, 
# in such chosen space generate a grid of 2 rows and 1 columm

ax2 = plt.subplot(subgs[0,0])
ax2.plot(data.flatten())
hel.labels(ax2, title = 'X-resolved trace', xlabel = 'abscissa', ylabel = 'ordinate')

ax3 = plt.subplot(subgs[1,0])
ax3.hist(data.flatten())
hel.labels(ax3, title = 'Distribution of a random sample', xlabel = 'values', ylabel = 'counts')

hel.savef(fig, fname, dpis, formats)
```
